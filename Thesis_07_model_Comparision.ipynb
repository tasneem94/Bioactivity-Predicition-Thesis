#Comparision of the models for Plasmodium Falciparum#

```python
import numpy as np
import matplotlib.pyplot as plt


# creating the dataset
data = {'FNN':97, 'RF':81, 'SVM':80}
courses = list(data.keys())
values = list(data.values())

fig = plt.figure(figsize = (10, 5))

# creating the bar plot
plt.bar(courses, values, color ='maroon',
		width = 0.4)

plt.xlabel('Models', fontweight ='bold', fontsize = 15)
plt.ylabel('Accuracy', fontweight ='bold', fontsize = 15)
plt.title("Model comparision for Plasmodium_falciparum")
plt.show()
```

#Model Performance Comparision for Plasmodium Falciparum VS Leukocyte Elastase# 

```python
import numpy as np
import matplotlib.pyplot as plt

# set width of bar
barWidth = 0.25
fig = plt.subplots(figsize =(12, 8))

# set height of bar
Plasmodium_Falciparum = [97, 81, 80]
Leukocyte_Elastase = [99,90,90]

# Set position of bar on X axis
br1 = np.arange(len(Plasmodium_Falciparum))
br2 = [x + barWidth for x in br1]

# Make the plot
plt.bar(br1, Plasmodium_Falciparum, color ='maroon', width = barWidth,
		edgecolor ='grey', label ='Plasmodium Falciparum')
plt.bar(br2, Leukocyte_Elastase, color ='pink', width = barWidth,
		edgecolor ='grey', label ='Leukocyte Elastase')


# Adding Xticks
plt.xlabel('Models', fontweight ='bold', fontsize = 15)
plt.ylabel('Accuracy', fontweight ='bold', fontsize = 15)
plt.xticks([r + barWidth for r in range(len(Plasmodium_Falciparum))],
		['FNN', 'RF', 'SVM'])

plt.legend()
plt.show()
```

#Our Result vs Base Paper's Result comparison# 

```python
import numpy as np
import matplotlib.pyplot as plt

# set width of bar
barWidth = 0.25
fig = plt.subplots(figsize =(12, 8))

# set height of bar
Our = [97, 80]
Their = [90,92]

# Set position of bar on X axis
br1 = np.arange(len(Our))
br2 = [x + barWidth for x in br1]

# Make the plot
plt.bar(br1, Our, color ='maroon', width = barWidth,
		edgecolor ='black', label ='Our Result')
plt.bar(br2, Their, color ='grey', width = barWidth,
		edgecolor ='black', label ='Their Result')


# Adding Xticks
plt.xlabel('Models', fontweight ='bold', fontsize = 15)
plt.ylabel('Accuracy', fontweight ='bold', fontsize = 15)
plt.xticks([r + barWidth for r in range(len(Our))],
		['FNN', 'SVM'])

plt.legend()
plt.show()
```
